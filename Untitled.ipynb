```python
import openai
import config

openai.api_key = config.apiKey
openai.api_base = "https://openrouter.ai/api/v1"
```

```python
def generateText(prompt, maxTokens, temp):
    res = openai.Completion.create(
        model="openai/gpt-4",
        prompt=prompt,
        max_tokens=maxTokens,
        temperature=temp #randomness, 1 => most random
    )
    #print(res)
    return f"{prompt} {res.choices[0].text.strip()}"
```

```python
generateText("Once upon a time", 50, 0)
```

```python
generateText("Once upon a time", 50, 1)
```
