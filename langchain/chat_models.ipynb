```python
from langchain_openai.chat_models import ChatOpenAI
import sys
sys.path.append("..")
import config

chat = ChatOpenAI(
    model = "openai/gpt-4",
    seed = 365,
    api_key=config.apiKey2,
    temperature = 0,
    base_url="https://openrouter.ai/api/v1"
)
```

#### System and human messages

```python
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
messageS = SystemMessage(content = """You are an chatbot as a creative programmer with a lot of experience in game dev, 
web dev and data science that answers questions with a wish to improve the questioner.""")
messageH = HumanMessage(content = "Recently I've been interested in game development. Can you suggest some begginer projects in Unity?")
```

```python
res = chat.invoke([messageS, messageH])
```

```python
res
```

```python
print(res.content)
```

```python
# Few shot prompting => we give a model some examples responses to learn from
from langchain_core.messages import AIMessage
messageHDog = HumanMessage(content = "I've recently adopted a dog. Can you suggest some dog names?")
# sarcastic response
messageAIDog = AIMessage(content = "Oh, absolutely. Because nothing screams 'I am a responsible pet owner' like asking a chatbot to name your new furball. How about 'Bark Twain' (if it is a literary hound)?")
messageHCat = HumanMessage(content = "I've recently adopted a cat. Can you suggest some cat names?")
messageAICat = AIMessage(content = "Oh, absolutely. Because nothing screams 'I am a unique and creative individual' like asking a chatbot to name your new cat. How about 'Furry McFurFace' or 'Catashtrophe'?")
messageHFish = HumanMessage(content = "I've recently adopted a fish. Can you suggest some fish names?")

res = chat.invoke([messageHDog, messageAIDog, messageHCat, messageAICat, messageHFish])
print(res.content)
```
