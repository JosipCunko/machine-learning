#### Templates

```python
from langchain_core.prompts import PromptTemplate
```

```python
TEMPLATE = """
System:
{description}

Human:
I've recently bought a new car: {car}.
Can you tell me some properties of {car}?
"""
```

```python
promptTemplate = PromptTemplate.from_template(template = TEMPLATE)
```

```python
promptTemplate
```

```python
promptValue = promptTemplate.invoke({"description": "The chatbot should provide a detailed information about every major property of a specific car.",
                                     "car": "Porsche 911 turbo S"})
```

```python
promptValue
```

#### Chat prompt templates

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts.chat import SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate
import sys
sys.path.append("..")
import config
chat = ChatOpenAI(
    model = "openai/gpt-4",
    seed = 365,
    api_key=config.apiKey2,
    max_tokens = 250,
    temperature = 0,
    base_url="https://openrouter.ai/api/v1"
)
```

```python
TEMPLATE_SY = "{description}"
TEMPLATE_H = "I've recently bought a new car: {car}. Can you tell me some properties of {car}?"
```

```python
messageTemplateSy = SystemMessagePromptTemplate.from_template(template = TEMPLATE_SY)
messageTemplateH = HumanMessagePromptTemplate.from_template(template = TEMPLATE_H)
print(messageTemplateH, messageTemplateSy )
```

```python
chatTemplate = ChatPromptTemplate.from_messages([messageTemplateSy, messageTemplateH])
chatTemplate
```

```python
chatValue = chatTemplate.invoke({"description": "The chatbot should provide a detailed but short information about some properties of a specific car. The properties are: engine, performance and exterior design.",
                                "car": "Rimac Nevera "})
```

```python
chatValue
```

```python
res = chat.invoke(chatValue) # accepts also instances from PromptValue class
```

```python
res
```

```python
print(res.content)
```

#### Few shot chat message prompt templates

```python
from langchain_core.prompts import FewShotChatMessagePromptTemplate, AIMessagePromptTemplate
```

```python
TEMPLATE_AI = "{response}"
messageTemplateAI = AIMessagePromptTemplate.from_template(TEMPLATE_AI)
```

```python
exampleTemplate = ChatPromptTemplate.from_messages([messageTemplateH, messageTemplateAI]) # blueprint
```

```python
examples = [{"car": "911 turbo S", 
            "response": """Sure, the Porsche 911 Turbo S is a high-performance sports car with impressive properties.\n\nEngine: The 911 Turbo S is powered by a 3.8-liter six-cylinder twin-turbo engine. This engine is capable of producing an impressive 640 horsepower and 590 lb-ft of torque. 
             It's paired with an eight-speed dual-clutch automatic transmission.\n\nPerformance: The car can accelerate from 0 to 60 mph in just 2.6 seconds, making it one of the fastest cars in its class. It has a top speed of 205 mph. The 911 Turbo S also features Porsche's Active Suspension Management system, 
             which adjusts the damping force on each wheel based on current road conditions and driving style.\n\nExterior Design: The Porsche 911 Turbo S boasts a sleek and aerodynamic design. It features wide rear arches and a large rear wing for improved downforce. The car also has LED matrix headlights and 20-inch front and 21-inch rear wheels. 
             The exterior design is not only stylish but also functional, improving the car's performance and handling."""
            },
           {"car": "Rimac Nevera",
            "response": """"Sure, the Rimac Nevera is a high-performance electric hypercar that boasts impressive properties in terms of engine, performance, and exterior design.\n\nEngine: The Nevera is powered by four independent permanent-magnet electric motors placed at each wheel. The total power output is a staggering 1,914 horsepower and 2,360 Nm of torque,
            making it one of the most powerful production cars in the world.\n\nPerformance: The Nevera is incredibly fast, with a top speed of 412 km/h (256 mph). It can accelerate from 0-60 mph in just 1.85 seconds, and 0-100 km/h in 1.97 seconds. The car's performance is further enhanced by its advanced torque vectoring system, which distributes power to the wheels 
            in a way that maximizes grip and stability.\n\nExterior Design: The Nevera's exterior design is both sleek and aerodynamic. It features a full carbon-fibre monocoque chassis and body panels, which not only look stunning but also contribute to the car's lightweight and high strength. The car's aerodynamics are actively 
            controlled, including the rear wing, underbody flaps, and bonnet turning vane, allowing it to adapt to different driving"""
            }
           ]
```

```python
fewShotPrompt = FewShotChatMessagePromptTemplate(examples = examples, example_prompt = exampleTemplate)
```

```python
chatTemplate = ChatPromptTemplate.from_messages([fewShotPrompt, messageTemplateH])
```

```python
chatValue = chatTemplate.invoke({"car": "Pagani Zonda R"})
```

```python
chatValue
```

```python
res = chat.invoke(chatValue)
```
