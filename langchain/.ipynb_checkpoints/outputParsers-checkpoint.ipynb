```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage
import sys
sys.path.append("..")
import config

chat = ChatOpenAI(
    model = "openai/gpt-4",
    seed = 365,
    api_key=config.apiKey2,
    temperature = 0,
    max_tokens = 100,
    base_url="https://openrouter.ai/api/v1"
)
```

#### String output parser

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
messageH = HumanMessage(content = "Can you give me an insteresting fact about parfumes that I didnt know about?")
```

```python
res = chat.invoke([messageH])
res
```

```python
strOutputParser = StrOutputParser()
```

```python
resParsed = strOutputParser.invoke(res)
resParsed
```

#### Comma seperated list output parser

```python
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
messageH = HumanMessage(content = f''' Can you suggest some names for a newborn baby?
{CommaSeparatedListOutputParser().get_format_instructions()}
''')
res = chat.invoke([messageH])
```

```python
listOutParser = CommaSeparatedListOutputParser()
resParsed = listOutParser.invoke(res)
resParsed
```

```python
print(listOutParser.get_format_instructions())
```
