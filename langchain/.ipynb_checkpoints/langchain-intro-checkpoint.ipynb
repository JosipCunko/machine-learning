```python
import os
#for key, value in os.environ.items():
   #print(f"{key}: {value}")
#os.environ["OPENAI_API_KEY] = ""
#%load_ext dotenv
#%dotenv

import openai
import config
```

```python
client = openai.OpenAI(
    api_key=config.apiKey2, #os.getenv("OPENAI_API_KEY")
    base_url="https://openrouter.ai/api/v1"
)
```

```python
completion = client.chat.completions.create(
    model =  "openai/gpt-4",
    messages = [
        { "role": "system", "content": "You are Bebe, a chatbot that reluctantly answers the questions in a sarcastic way."},
        { "role": "user", "content": "I've recently adopted a dog. Can you suggest some dog names?"}],
    
)
completion
```

```python
completion = client.chat.completions.create(
    model =  "openai/gpt-4",
    messages = [
        { "role": "system", "content": "You are Bebe, a chatbot that reluctantly answers the questions in a sarcastic way."},
        { "role": "user", "content": "Explain what a black hole is"}],
    max_tokens = 250,
    temperature = 2,
    seed = 365
)
completion.choices[0].message.content
```

```python
completion = client.chat.completions.create(
    model =  "openai/gpt-4",
    messages = [
        { "role": "system", "content": "You are Bebe, a chatbot that reluctantly answers the questions in a sarcastic way."},
        { "role": "user", "content": "Explain how a buck converter works"}],
    max_tokens = 300,
    temperature = 0,
    seed = 300,
    stream = True
)
completion
```

```python
for i in completion:
    print(i.choices[0].delta.content, end = "")
```
